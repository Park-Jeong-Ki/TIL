```python
# 알파벳 찾기
# Word = 입려 문자
# result_dict = 답이 될 dict

Word = str(input())
result_dict = {}

#result_dict 초기화
for i in range(97,123):
  result_dict[chr(i)] = -1

#result_dict 값 변경
for i in range(len(Word)):
  if result_dict[Word[i]]== -1:
    result_dict[Word[i]] =i

for k,v in result_dict.items():
  print(v, end = ' ')
```

```python
# 문자열 반복

# N에 몇번의 값을 받을지 저장
N = int(input())
result_dict = {}

for i in range(N):
  ans_list=[]
  num, Word = map(str,input().split())
  for s in list(Word):
    ans_list.append(s*int(num))
  ans_list = ''.join(ans_list)

  result_dict[i] = ans_list

for k,v in result_dict.items():
  print(v)
```

```python
#단어 공부

Word = str(input()).upper()
result_dict = {}
#result_dict 초기화
for i in range(65,91):
  result_dict[chr(i)] = 0

for s in list(Word):
  result_dict[s] +=1

max_val = 0
max_key = None
for k,v in result_dict.items():
  if v > max_val:
    max_val = v
    max_key = k

for k,v in result_dict.items():
  if k != max_key and v == max_val:
    max_val = '?'
    max_key = '?'

if max_val == '?':
  print('?')
else:
  dic_max = max(result_dict, key=result_dict.get)
  print(dic_max)
```

```python
# 다이얼

Word = str(input())

sec_dict = {}

for i in range(1,11):
  if i ==10:
    sec_dict[0]=1+i
  else:
   sec_dict[i]=1+i

ABC_dict = {}

for i in range(65,91):
  if 65 <= i <= 67:
    ABC_dict[chr(i)] = 2
  elif 68 <= i <= 70:
    ABC_dict[chr(i)] = 3
  elif 71 <= i <= 73:
    ABC_dict[chr(i)] = 4
  elif 74 <= i <= 76:
    ABC_dict[chr(i)] = 5
  elif 77 <= i <= 79:
    ABC_dict[chr(i)] = 6
  elif 80 <= i <= 83:
    ABC_dict[chr(i)] = 7
  elif 84 <= i <= 86:
    ABC_dict[chr(i)] = 8
  elif 87 <= i <= 90:
    ABC_dict[chr(i)] = 9

Word_list = list(Word)

sec_sum = 0
for w in Word_list:
  sec_sum += sec_dict[ABC_dict[w]]
print(sec_sum)
```

```python
## 그룹 단어 체커

N = int(input())
word_dict={}
word_list=[]
pre_str = None
group = None
cnt = 0
for i in range(N):
  group=None
  pre_str=None

  Word = str(input())
  word_list.append(Word)
  Word_to_str = list(Word)

  for w in Word_to_str:
    word_dict[w] = 1

  for s in Word_to_str:
      if word_dict[s] == 'O':
        group = False
        cnt +=0
        break
      elif pre_str != s:
        word_dict[pre_str] = 'O'
        pre_str = s
        group = True
      elif pre_str == s:
        pre_str = s
        group = True
  if group == True:
    cnt +=1
  else:
    cnt +=0
print(cnt)
```
