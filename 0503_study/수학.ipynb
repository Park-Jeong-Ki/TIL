```python
# 분수찾기

# 분모만 쳐보면 1   2 1   1 2 3   4 3 2 1   1 2 3 4 5   6 5 4 3 2 1 

# 분자만 쳐보면 1   1 2   3 2 1   1 2 3 4  5 4 3 2 1   1 2 3 4 5 6

# 즉 둘이 반대임


# 만약 5가 들어가면 1+2+3 >5 그래서 3번째 6-5 =1  뒤에서 인덱스 -1-1?
# 민약 7이 들어가면 1+2+3+4 >7 10-7 = 3 뒤에서 인덱스 [-3-1]
# 만약 11이 들어가면 1+2+3+4+5 >11 15-11 = 4 뒤에서 인덱스 [-4-1]


# 인덱스로 하지말고 그냥 빼면됨
# 

N = int(input())
bunza = 0
num_list = []

num = 0
sum = 0

while sum < N:
  num += 1
  sum += num

sum = sum - num   # 이 대각선 위에 존재

if num%2 ==0: # 짝수일때 분자가 큰수부터 시작
  bunza = N - sum
  bunmo = num - bunza +1  # bunmo +bunza = num 항상 값이 같음 분모분자 더한거

else:
  bunza = num-(N-sum) +1
  bunmo = N - sum

print(f'{bunza}/{bunmo}')
  
```

```python
# ACM호텔

N = int(input())
ans = []
def get_ans():
  H,W,N = map(int,input().split())
  if N%H != 0:
    ho = N//H +1
  else:
    ho = N//H
  floor = N%H
  if floor ==0:
    floor = H
  return floor,ho

for i in range(N):
  f,h = get_ans()
  if h<10:
    ans.append(f'{f}0{h}')
  else:
    ans.append(f'{f}{h}')
for a in ans:
  print(a)
```

```python
# 부녀회장이 될테야


# 2층의 3호에 살려면

# 1층 1 2 3 다 더한거

# 1층의 1은 1
# 1층의 2는 1+2
# 1층의 3은 1+2+3

#2층의 3호는 1 + 1+2  1+2+3

t = int(input())

for i in range(t):
  floor = int(input()) # 층
  ho = int(input()) # 호
  F0 = [i for i in range(1,ho+1)] # 0층 초기화
  for i in range(floor):
    for j in range(1,ho):
      F0[j] += F0[j-1] # 층별 호실 사람수 더해주기
  print(F0[-1])
```
