```python
import numpy as np
import pandas as pd
import openpyxl

df1 = pd.read_excel('test.xlsx', sheet_name="Sheet1")
df2 = pd.read_excel('test.xlsx', sheet_name='Sheet2')

df1_copy = df1.copy()

def change_df (c_list, df1_copy):
    for col in c_list:
        for j, a in enumerate(df1_copy[col]):
            for i, name in enumerate(df2['이름']):
                temp_index = "NaN"
                if a == "NaN":
                    pass
                elif a == name :
                    temp_index = i + 1
                    break
            df1_copy[col][j] = temp_index
    return df1_copy
c_list = list(df1.columns)
change_df(c_list, df1_copy)

df1_copy.columns = ['b1', 'b2', 'b3', 'b4', 'b5']

result = pd.concat([df1, df1_copy], axis=1)
result
    
```
